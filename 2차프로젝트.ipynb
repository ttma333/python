 ### 2년간 매출순위 높은 점포순 Top 20

```python
zumpo=data_5['점포코드'].unique()
```

```python
zumpo=list(zumpo)
```

```python
zumpo_year = []
for i in zumpo:
  a=data_5[data_5['점포코드']==i]['구매금액'].sum()
  zumpo_year.append(a)
```

```python
총구매={'점포코드' : zumpo ,'총구매금액' : zumpo_year}
_2year_zumpo_total=pd.DataFrame(총구매)
```

```python
zumpo_2year=_2year_zumpo_total.sort_values('총구매금액',ascending=False)
zumpo_2year.head(20)
```

### 2014년 매출높은 점포순 top 20

```python
year_2014=data_5[data_5['구매일자'] < 20150101]
```

```python
year_2014_list=year_2014['점포코드'].unique()
```

```python
year_2014_list=list(year_2014_list)
```

```python
year_2014_list2 = []
for i in year_2014_list:
  a=year_2014[year_2014['점포코드']==i]['구매금액'].sum()
  year_2014_list2.append(a)
```

```python
총구매={'점포코드' : year_2014_list ,'총구매금액' : year_2014_list2}
year_2014_zumpo=pd.DataFrame(총구매)
```

```python
year_2014_zumpo=year_2014_zumpo.sort_values('총구매금액',ascending=False)
year_2014_zumpo.head(20)
```

### 2015년 매출 높은 점포순 top 20


```python
year_2015=data_5[data_5['구매일자'] > 20141231]
```

```python
year_2015_zumpo=year_2015_zumpo.reset_index()
```

```python
year_2015_zumpo=year_2015.groupby('점포코드')[['구매금액']].sum().sort_values('구매금액',ascending=False)
year_2015_zumpo.head(20)
```
